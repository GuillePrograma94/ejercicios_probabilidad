## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
# import numpy.
import numpy as np

#We determine the number of throws.
throws = 1000

#We create a function that will run as many times as n_throws.
def throw_both_dice(n):
    #We start counting how many times we obtain a number that is greater than 7 or an even number.
    times_result = 0
    #We generate a loop to throw the dice 1000 times.
    for i in range(n):
        #Results for each die separately.
        die_1 = np.random.randint(1,7)
        die_2 = np.random.randint(1,7)
        #Add the values of each die.
        dice_result = die_1 + die_2
        #Determine if the sum is greater than 7 or an even number.
        if dice_result > 7 or dice_result % 2 == 0:
            times_result += 1
    #Divide the number from step 2 by the number of iterations.
    return times_result / n

print(f'The probability to obtain a number that is either an even number or greater than 7 is:{throw_both_dice(throws)*100:2}')
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. Take all of the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
            
# We define the number of experiments:
n_experiments = 1000

# We will store the results in a list:
results = []


# We define the loop to draw the balls.
for _ in range(n_experiments):
    chosen_balls = np.random.choice(list(ball_box.values()), size=5, replace = True)
    results.append(chosen_balls)

# We count how many are 3 white and 2 red, and how many are all the same colour.
count_3_2 = 0
count_all = 0

# We loop through our results and count how many of them were successful.
for chosen_ball in results:
    #We count each colour:
    white_ = np.sum(chosen_ball == "White")    
    red_ = np.sum(chosen_ball == "Red")
    green_ = np.sum(chosen_ball == "Green")

    if white_ == 3 and red_ == 2:
        count_3_2 += 1

    elif white_ == 5 or red_ == 5 or green_ == 5:
        count_all += 1

#We calculate probabilities.
probability_3_2 = count_3_2 / n_experiments
probability_all = count_all / n_experiments

print(f"Probability of taking 3 white and 2 red: {probability_3_2 * 100:2}%")
print(f"Probability of taking all balls of the same color: {probability_all * 100:2}%")
```
